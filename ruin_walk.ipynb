```python
import numpy
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact
from collections import defaultdict
```

```python
#change default size of plots and font
plt.rcParams.update({'font.size': 22, 'figure.figsize': (14,8)})

coinflip = lambda: numpy.random.binomial(1,0.5)*2-1

def toRuin():
    k = 1
    walk = [k]
    while k>0:
        k += coinflip()
        walk.append(k)
    return walk

w = toRuin()
while len(w)<3:
    w = toRuin()
x = range(0, len(w))
plt.plot(x, w)
plt.show()
```

```python
def show(n):
    lengths = defaultdict(int)
    maximums = defaultdict(int)
    for i in range(n):
        walk = toRuin()
        length = len(walk)-1
        maximum = max(walk)
        lengths[length] += 1
        maximums[maximum] += 1
        if length > 1:
            x = range(0, len(walk))
            plt.plot(x, walk)
    
    longest = max(lengths.keys())
    rest = -longest
    for length, count in lengths.items():
        rest += length*count
    
    print('   longest walk:', longest)
    print('sum of the rest:', rest)
    plt.show()
```

```python
interact(show,\
         n = widgets.IntText(value=10,continuous_update=False)
        );
```
